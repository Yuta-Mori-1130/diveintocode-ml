# 【問題2】データの概観の把握

* データの取得

```python
#df = pd.read_csv('train.csv')
df_train = pd.read_csv('application_train.csv', index_col=0)
df_test = pd.read_csv('application_test.csv', index_col=0)
```

* データセット自体の調査

```python
print(df_train.shape)
df_train.head(5)
```

```python
print(df_test.shape)
df_test.head(5)
```

```python
# GrLivAreaとYearBuiltを抜き出す。
X = df_train[["AMT_INCOME_TOTAL", "AMT_CREDIT"]]
y = df_train["TARGET"]
```

```python
# 訓練データと検証データの分割

from sklearn.model_selection import train_test_split

X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 前処理
```

```python
# 学習と推定
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5) # インスタンス生成。n_neighbors:Kの数
knn.fit(X_train, y_train) # モデル作成実行

y_pred = knn.predict_proba(X_test) # 予測実行

print(y_pred.shape)
print(y_pred)
```

# 【問題3】テストデータに対する推定

```python
# テストデータを読み込み推定

X_test_eva = np.array(df_test[["AMT_INCOME_TOTAL", "AMT_CREDIT"]])
y_pred_eva = knn.predict_proba(X_test_eva) # 予測実行

print(X_test_eva.shape)
print(y_pred_eva)
```
