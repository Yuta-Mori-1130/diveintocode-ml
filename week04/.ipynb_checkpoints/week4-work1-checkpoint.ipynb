# 【問題1】
## コンペティション内容の確認
コンペティションのOverviewページ読み、「Home Credit Default Risk」について以下の観点について確認してください。

- 何を学習し、何を予測するのか
- どのようなファイルを作りKaggleに提出するか
- 提出されたものはどういった指標値で評価されるのか

## 何を学習し、何を予測するのか
ローンの情報(車を持ってるか、不動産を持ってるか、学歴など)を元に  
クライアントのローンの返済能力を予測する

## どのようなファイルを作りKaggleに提出するか
(TARGETは予測結果(%))

下記のような形式(csvファイル)
```
SK_ID_CURR,TARGET
100001,0.1
100005,0.9
100013,0.2
etc.
```

## 提出されたものはどういった指標値で評価されるのか
ROC曲線を用いて評価される

まずは単純な方法による **ベースラインモデル** を作成します。精度の基準となるモデルです。

精度を高くする必要はありません。エラーなく実行でき、Kaggle側に推定値を提出できることを目指します。

# 【問題2】
## 学習と検証
データを簡単に分析、前処理し、学習、検証するまでの一連の流れを作成・実行してください。

検証にはこのコンペティションで使用される評価指標を用いるようにしてください。学習に用いる手法は指定しません。

````python
import numpy as np
import pandas as pd
import seaborn as sns
import missingno as msno
import matplotlib
import matplotlib.pyplot as plt
````

````python
df_train = pd.read_csv('application_train.csv')
df_test = pd.read_csv('application_test.csv')
````

````python
print(df_train.shape)
df_train.head()
````

````python
print(df_test.shape)
df_test.head()
````

特徴量:"EXT_SOURCE_1"のみを利用してベースラインモデルを作成する

````python
from sklearn.model_selection import train_test_split

# 欠損値を平均値で補完
df_train_fillna = df_train.fillna({"EXT_SOURCE_1":df_train["EXT_SOURCE_1"].mean()})

# 型変換
X = df_train_fillna["EXT_SOURCE_1"].values
y = df_train_fillna["TARGET"].values

# X_train（学習用）とX_valid（検証用）に分割
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, random_state=0)
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
````

````python
from sklearn.ensemble import RandomForestClassifier

# 学習
clf = RandomForestClassifier().fit(X_train.reshape(-1, 1), y_train)

y_vaild_pred = clf.predict_proba(X_valid.reshape(-1, 1))
````

````python
from sklearn.metrics import roc_auc_score

# 評価
score_temp = roc_auc_score(y_valid, y_vaild_pred[:,1])
print(f'AUC_test: {score_temp}')
````

# 【問題3】
## テストデータに対する推定
テストデータ（`application_test.csv`）に対して推定を行い、  
Kaggleに提出を行ってください。

正しく提出が行えていれば、精度は低くても構いません。  
→Score 0.53376

````python
# 欠損値を確認
df_test[["SK_ID_CURR", "EXT_SOURCE_1"]].isnull().sum()
````

````python
# 欠損値を平均値で補完
df_test_fillna = df_test.fillna({"EXT_SOURCE_1":df_train["EXT_SOURCE_1"].mean()})

# 型変換
#ids = df_test_fillna['SK_ID_CURR'].values
X_test = df_test_fillna["EXT_SOURCE_1"].values

# 予測
y_test_pred = clf.predict_proba(X_test.reshape(-1, 1))
````

````python
# 提出用データの作成

# NumPy配列ndarrayからpandasに型変換する
df_score = pd.DataFrame(y_test_pred[:,1], columns=["TARGET"])
#print(df_score)

# 提出フォーマットのデータフレームを作成する
df_sk_id = df_test_fillna["SK_ID_CURR"]
#print(df_sk_id)

df_submission = pd.concat([df_sk_id, df_score], axis=1)
display(df_submission)

# Score 0.53376
df_submission.to_csv('submission.csv', index=False)
````

## パターン1（スコア:0.61990）

[kaggle:EDA](https://www.kaggle.com/osciiart/homecreditrisk-extensive-eda-baseline-model-jp)

上記のEDAから"EXT_SOURCE_1,2,3"が重要だということなのでこれを利用する  

````python
# 欠損値を確認
df_train[['SK_ID_CURR', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].isnull().sum()
````

````python
# 欠損値を平均値で補完
df_train_fillna = df_train.fillna({
                                "EXT_SOURCE_1":df_train["EXT_SOURCE_1"].mean(),
                                "EXT_SOURCE_2":df_train["EXT_SOURCE_2"].mean(),
                                "EXT_SOURCE_3":df_train["EXT_SOURCE_3"].mean(),
                                })

df_test_fillna = df_test.fillna({
                                "EXT_SOURCE_1":df_train["EXT_SOURCE_1"].mean(),
                                "EXT_SOURCE_2":df_train["EXT_SOURCE_2"].mean(),
                                "EXT_SOURCE_3":df_train["EXT_SOURCE_3"].mean(),
                                })
````

````python
#df_train_fillna[['SK_ID_CURR', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].isnull().sum()
````

````python
# 前処理
from sklearn.model_selection import train_test_split

# 特徴量（説明変数）をX、正解（目的変数）をyというndarrayに格納
X = df_train_fillna[["EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3"]].values
y = df_train_fillna["TARGET"].values

# X_train（学習用）とX_valid（検証用）に分割
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, random_state=0)
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
````

````python
from sklearn.ensemble import RandomForestClassifier

# 学習
clf = RandomForestClassifier().fit(X_train, y_train)

y_vaild_pred = clf.predict_proba(X_valid)
````

````python
from sklearn.metrics import roc_auc_score

# 評価
score_temp = roc_auc_score(y_valid, y_vaild_pred[:,1])
print(f'AUC_test: {score_temp}')
````

````python
# テストデータに対する推定
# 型変換
X_test = df_test_fillna[["EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3"]].values

# 予測
y_test_pred = clf.predict_proba(X_test)

# 提出用データの作成
# NumPy配列ndarrayからpandasに型変換する
df_score = pd.DataFrame(y_test_pred[:,1], columns=["TARGET"])
#print(df_score)

# 提出フォーマットのデータフレームを作成する
df_sk_id = df_test_fillna["SK_ID_CURR"]
#print(df_sk_id)

df_submission = pd.concat([df_sk_id, df_score], axis=1)
display(df_submission)

# Score 0.61990
df_submission.to_csv('submission.csv', index=False)
````
