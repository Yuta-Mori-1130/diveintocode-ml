# 【問題3】手法の比較

```python
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# 任意のモデルで学習し、プロットする関数。また、評価結果を返す関数。

def calc(models, X_train, y_train):
    
    # 学習
    clf =  models()
    clf.fit(X_train, y_train)
        
    # 推定
    y_pred = clf.predict(X_test)
    # print(y_pred.shape) # (365,)

    # 評価（平均二乗誤差（標準偏差、ばらつき、MSE））
    score = mean_squared_error(y_test, y_pred)
    print("平均二乗誤差（標準偏差、ばらつき、MSE）", score)
    
    # グラフの描画
    plt.title("scatter", fontsize=20) # (5)タイトル
    plt.xlabel("GrLivArea", fontsize=20) # (6)x軸ラベル
    plt.ylabel("SalePrice", fontsize=20) # (7)y軸ラベル

    plt.scatter(X_test["GrLivArea"], y_pred, s=10, c="g", marker="D", label="virgiclor(1)") #(10)散布図の描画
    plt.show()

    # グラフの描画
    plt.title("scatter", fontsize=20) # (5)タイトル
    plt.xlabel("YearBuilt", fontsize=20) # (6)x軸ラベル
    plt.ylabel("SalePrice", fontsize=20) # (7)y軸ラベル

    plt.scatter(X_test["YearBuilt"], y_pred, s=10, c="b", marker="D", label="virgiclor(1)") #(10)散布図の描画
    plt.show()
    
    return score
```

```python
print("線形回帰")
score0 = calc(LinearRegression, X_train, y_train)
```

```python
print("SVM")
score1 = calc(SVR, X_train, y_train)
```

```python
print("決定木")
score2 = calc(DecisionTreeRegressor, X_train, y_train)
```

```python
print("ランダムフォレスト")
score3 = calc(RandomForestRegressor, X_train, y_train)
```
