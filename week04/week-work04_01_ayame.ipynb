```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
```

[データセット](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html)

sepal:がく  
petal:花弁

```python
# データの読み込み
data_iris = load_iris()

X = pd.DataFrame(data_iris.data, columns=["sepal_length","sepal_width","petal_length","petal_width"])
#X.head(5)

y = pd.DataFrame(data_iris.target, columns=["Species"])
#y.head(5)

df = pd.concat([X, y], axis=1)
df.head(5)
```

```python
# データフレームからこれらを抜き出してください。
# virgicolorとvirginica
# sepal_lengthとpetal_length
# 参考情報:setosa:0, virgiclor:1, virginica:2

# virgicolorとvirginicaを抜き出す。
df_temp1 = df[ df["Species"] == 1 ]
df_temp2 = df[ df["Species"] == 2 ]
df_temp3 = pd.concat([df_temp1, df_temp2], axis=0)

# sepal_lengthとpetal_lengthを抜き出し結合する
df2 = df_temp3[["sepal_length", "petal_length", "Species"]]
#print(df2)
print(df2.head(5))
print(df2.tail(5))
```

# 【問題3】前処理・訓練データと検証データの分割

* 訓練データと検証データの分割を行います。  
訓練データ75%、検証データ25%として分割してください。

```python
from sklearn.model_selection import train_test_split

# 特徴量（説明変数）をX、正解（目的変数）をyというndarrayに格納
X = df2[["sepal_length", "petal_length"]].values
#print(X)
#print(type(X))

y = df2[["Species"]].values
#print(y)
#print(type(y))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
#print(X_train)
#print(y_train.shape)
#print(y_train)
#print(X_test)
```

# 【問題4】前処理・標準化

* 分類精度をあげるための加工を行います。手法ごとに適切な方法は異なりますが、ここでは一例として 標準化 を行うことにします。

```python
# 標準化は訓練用のデータでfitを実施
# 訓練用、検証用双方のデータにtransformを行います

# 良く使うメソッドは次の3つ。
# fit(X)
# 配列Xの平均と分散を計算して、記憶する（変換は行わない）。

#transform(X)
#配列Xに変換を施して、変換後の配列を返す。

#fit_transform(X)
#配列Xに対して、fitとtransformを同時に行う。

#なお、2次元配列を変換する場合、縦 (axis=0) 方向に変換が行われる。
```

```python
from sklearn.preprocessing import StandardScaler

print("標準化前")
print(X_train.mean(axis=0))
print(X_train.std(axis=0))

scaler = StandardScaler()
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)

print("標準化後")
print(X_train_std.mean(axis=0))
print(X_train_std.std(axis=0))

X_test_std = scaler.transform(X_test)
```

# 【問題5】学習と推定

```python
# 最近傍法（k-nearest neighbors, k-nn）で学習
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

#print(y_train.shape)
#print(y_train.reshape(75,))

for i in range(1, 6):
    #print(i)
    knn = KNeighborsClassifier(n_neighbors=i) # インスタンス生成。n_neighbors:Kの数
    knn.fit(X_train_std, y_train.reshape(75,)) # モデル作成実行
    y_pred = knn.predict(X_test_std) # 予測実行
    print(y_pred)
    #print(i,"-nn", " = ",accuracy_score(y_test, y_pred))
```

# 【問題6】評価

* scikit-learnを使い4種類( Accuracy（正解率）、Precision（適合率）、Recall（再現率）、F値)の指標を計算する

```python
print("Accuracy（正解率）")

for i in range(1, 6):
    #print(i)
    knn = KNeighborsClassifier(n_neighbors=i) # インスタンス生成。n_neighbors:Kの数
    knn.fit(X_train_std, y_train.reshape(75,)) # モデル作成実行
    y_pred = knn.predict(X_test_std) # 予測実行
    print(i,"-nn", " = ",accuracy_score(y_test, y_pred))

# 問8用の表示用にデフォルト5-nnのデータを保存
list_score_KN = []
list_score_KN.append(accuracy_score(y_test, y_pred))
```

```python
print("Precision（適合率）")

for i in range(1, 6):
    #print(i)
    knn = KNeighborsClassifier(n_neighbors=i) # インスタンス生成。n_neighbors:Kの数
    knn.fit(X_train_std, y_train.reshape(75,)) # モデル作成実行
    y_pred = knn.predict(X_test_std) # 予測実行
    print(i,"-nn", " = ",precision_score(y_test, y_pred))

# 問8用の表示用にデフォルト5-nnのデータを保存
list_score_KN.append(precision_score(y_test, y_pred))
```

```python
print("Recall（再現率）")

for i in range(1, 6):
    #print(i)
    knn = KNeighborsClassifier(n_neighbors=i) # インスタンス生成。n_neighbors:Kの数
    knn.fit(X_train_std, y_train.reshape(75,)) # モデル作成実行
    y_pred = knn.predict(X_test_std) # 予測実行
    print(i,"-nn", " = ",recall_score(y_test, y_pred))

# 問8用の表示用にデフォルト5-nnのデータを保存
list_score_KN.append(recall_score(y_test, y_pred))
```

```python
print("F値")

for i in range(1, 6):
    #print(i)
    knn = KNeighborsClassifier(n_neighbors=i) # インスタンス生成。n_neighbors:Kの数
    knn.fit(X_train_std, y_train.reshape(75,)) # モデル作成実行
    y_pred = knn.predict(X_test_std) # 予測実行
    print(i,"-nn", " = ",f1_score(y_test, y_pred))
    
# 問8用の表示用にデフォルト5-nnのデータを保存
list_score_KN.append(f1_score(y_test, y_pred))
```

```python
print("混同行列")

for i in range(1, 6):
    #print(i)
    knn = KNeighborsClassifier(n_neighbors=i) # インスタンス生成。n_neighbors:Kの数
    knn.fit(X_train_std, y_train.reshape(75,)) # モデル作成実行
    y_pred = knn.predict(X_test_std) # 予測実行
    print(i,"-nn", " = ")
    print(confusion_matrix(y_test, y_pred))
```

* 4種類の指標値の意味について簡単に調査して文章でまとめる

Accuracy（正解率）:正や負と予測したデータのうち，実際にそうであるものの割合  
Precision（適合率）:正と予測したデータのうち，実際に正であるものの割合  
Recall（再現率）:実際に正であるもののうち，正であると予測されたものの割合    
F値:再現率と適合率の調和平均．  

TP(True Positive):正しいものを正しく推測できたもの  
TN(True Negative):正しくないものを正しく推測できたもの  
FN(False Negaitve):正しいものを誤って正しいと推測したもの  
FP(False Positive):正しくないものを誤って正しいと推測したもの  

混合行列 =   
[TP, FN]  
[FP, TN]  

# 【問題7】可視化

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2
    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)
    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=5) # インスタンス生成。n_neighbors:Kの数
knn.fit(X_train_std, y_train.reshape(75,))

decision_region(X_train_std, y_train.reshape(75,), knn, step=0.01, title='decision region of 5-nn', xlabel='sepal_length', ylabel='petal_length', target_names=['versicolor', 'virginica'])
```
