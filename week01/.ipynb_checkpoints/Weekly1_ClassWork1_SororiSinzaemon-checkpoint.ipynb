# 【問題1】100日目の米粒の数

100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。
また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。
（縦軸が米の数、横軸が日数）

```python
NUM_DAYS = 100

list_num_day = list( range(1,NUM_DAYS+1) )

list_num_rice = [] #その日にもらう米の数のリスト
list_num_sum_rice = [] # その日までに累計で貰う米の数のリスト

#print(list_num_day)
#print(len(list_num_day))

num_temp_rice = 0 # その日にもらう米の数
num_temp_sum_rice = 0 # その日までに累計で貰う米の数

for i in list_num_day:
    num_temp_rice = 2**(i-1)
    num_temp_sum_rice = num_temp_sum_rice + num_temp_rice
    
    #print(i, num_temp_rice, num_temp_sum_rice)

    list_num_rice.append(num_temp_rice)
    list_num_sum_rice.append(num_temp_sum_rice)
    
#print(list_num_rice)
#print(len(list_num_rice))

#print(list_num_sum_rice)
#print(len(list_num_sum_rice))

"""
グラフを表示する。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("day")
plt.ylabel("The number of rice you get on that day")
plt.plot(list_num_day, list_num_rice) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
"""
グラフを表示する。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("day")
plt.ylabel("The total number of rice received by that day")
plt.plot(list_num_day, list_num_sum_rice) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

# 【問題2】100日目以外の米粒の数

この逸話は30日だったり81日だったりと日数が曖昧です。
どの話にも対応できるように日数を引数とした 関数 を作成しましょう。
関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。
前の問題同様に折れ線グラフも出力してください。

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    input 
        day: int 米を貰う日数 (default : 100)
    
    output
        list_num_day 日付情報（リストの長さはdayに等しくなる）
        list_num_rice ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
        list_num_sum_rice ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    # ここにコードを書く。passは何もしないことを意味する命令である。
    list_num_day = list( range(1,day+1) )

    list_num_rice = [] #その日にもらう米の数のリスト
    list_num_sum_rice = [] # その日までに累計で貰う米の数のリスト

    num_temp_rice = 0 # その日にもらう米の数
    num_temp_sum_rice = 0 # その日までに累計で貰う米の数

    for i in list_num_day:
        num_temp_rice = 2**(i-1)
        num_temp_sum_rice = num_temp_sum_rice + num_temp_rice

        list_num_rice.append(num_temp_rice)
        list_num_sum_rice.append(num_temp_sum_rice)

    return list_num_day, list_num_rice, list_num_sum_rice

# コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。
# 90日とする
list_n_day, list_n_grains, list_total_grains = compute_sorori_shinzaemon(day= 90)

#print(list_n_day)
#print(list_n_grains)
#print(len(list_n_grains))
#print(list_total_grains)
#print(len(list_total_grains))
```

```python
"""
グラフを表示する。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("day")
plt.ylabel("The number of rice you get on that day")
plt.plot(list_n_day, list_n_grains) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
"""
グラフを表示する。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("day")
plt.ylabel("The total number of rice received by that day")
plt.plot(list_n_day, list_total_grains) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

# 【問題3】何人が何日生活できる米か

この米があれば何人が何日生活できるか計算する関数を作成してください。
コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。

関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。

ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
def calc_rice_living_days(num_rice = 6000 ,num_people = 1):
    """
    N粒の米で、M人がQ日生活できるかを求める関数。

    input:
        米粒の数(int) と 人数(int)

    output
        出力は 生活できる日数(int) 

    参照:https://suiteidan.com/suitei-3/
        1日に必要な米粒の数
        1日に１合必要で、1合に入っている米粒の数 6000［粒］
    """
    NUM_RICE_REQUIRED = 6000
    
    # 小数点以下を切り捨てし、整数値を返す
    num_living_days = int(num_rice/(num_people * NUM_RICE_REQUIRED ))
    
    return num_living_days

#arg1 = 500000
#print(list_total_grains)
#print(list_total_grains[-1])
arg1 = list_total_grains[-1]
arg2 = 15
ans1 = calc_rice_living_days(arg1, arg2)
print("{}粒で{}人の場合、{}日生活できます。".format(arg1, arg2, ans1))
```
