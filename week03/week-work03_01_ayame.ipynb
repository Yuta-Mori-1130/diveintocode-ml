# 【問題6】図の作成

* ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

```python
#print(df["Species"])

count_label1 = len(df[ df["Species"] == 0 ])
count_label2 = len(df[ df["Species"] == 1 ])
count_label3 = len(df[ df["Species"] == 2 ])

print("setosa", count_label1)
print("virgiclor", count_label2)
print("virginica", count_label3)

label = ["setosa(0)", "virgiclor(1)", "virginica(2)"]

x = np.array([count_label1,count_label2,count_label3])
plt.pie(x, labels=label, counterclock=False, startangle=90, autopct="%1.1f%%")

plt.axis('equal')
```

* 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

```python
data1 = df[ df["Species"] == 0 ]
data2 = df[ df["Species"] == 1 ]
data3 = df[ df["Species"] == 2 ]

for column_name in df:
    #print(column_name)
    if column_name == "Species":
        # 最終列(ラベル)は表示しない。
        break
    
    data = (data1[column_name], data2[column_name], data3[column_name])

    fig, axes = plt.subplots(figsize = (12, 8))
    plt.boxplot(data, patch_artist=True, widths=0.2,boxprops=dict(facecolor='skyblue',
                             color='black', linewidth=1))
    axes.set_xticklabels(['setosa', 'virgiclor', "virginica"])
    #plt.title('Box plot')
    #plt.xlabel('x')
    plt.ylabel(column_name)
    plt.show()
```

* 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

```python
import seaborn as sns

for column_name in df:
    #print(column_name)
    if column_name == "Species":
        # 最終列(ラベル)は表示しない。
        break

    plt.style.use('default')
    sns.set()
    sns.set_style('whitegrid')
    sns.set_palette('gray')

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

    ax.violinplot([data1[column_name], data2[column_name], data3[column_name]])
    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(["setosa", "virgiclor", "virginica"])
    #ax.set_xlabel('x')
    ax.set_ylabel(column_name)

    plt.show()
```

* 箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。

* 箱ひげ図: データの四分位点が分かるもの  
→平均値などがわかる。
* バイオリン図:データの分布が分かるもの  
→データの分布がわかる

# 【問題7】特徴量間の関係の確認

* 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
#count_label1 = len(df[ df["Species"] == 0 ])
#count_label2 = len(df[ df["Species"] == 1 ])
#count_label3 = len(df[ df["Species"] == 2 ])

#data1 = df["sepal_length"]
#data2 = df["sepal_width"]
#data3 = df["petal_length"]
#data4 = df["petal_width"]

# label = ["setosa(0)", "virgiclor(1)", "virginica(2)"]

#print(df[df["Species"] == 0])

# データ準備

x0 = df[df["Species"] == 0]["sepal_length"]
y0 = df[df["Species"] == 0]["sepal_width"]

x1 = df[df["Species"] == 1]["sepal_length"]
y1 = df[df["Species"] == 1]["sepal_width"]

x2 = df[df["Species"] == 2]["sepal_length"]
y2 = df[df["Species"] == 2]["sepal_width"]

# グラフの装飾
plt.xlim(0, 10) # (3)x軸の表示範囲
plt.ylim(0, 10) # (4)y軸の表示範囲
plt.title("scatter", fontsize=20) # (5)タイトル
plt.xlabel("sepal_length", fontsize=20) # (6)x軸ラベル
plt.ylabel("sepal_width", fontsize=20) # (7)y軸ラベル
plt.grid(True) # (8)目盛線の表示
plt.tick_params(labelsize = 12) # (9)目盛線のラベルサイズ
 
# グラフの描画
plt.scatter(x0, y0, s=10, c="r", marker="D", label="setosa(0)") #(10)散布図の描画
plt.scatter(x1, y1, s=10, c="g", marker="D", label="virgiclor(1)") #(10)散布図の描画
plt.scatter(x2, y2, s=10, c="b", marker="D", label="virginica(2)") #(10)散布図の描画

plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=18)

plt.show()
```

* 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

[参考情報](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

[参考情報](https://note.nkmk.me/python-seaborn-pandas-pairplot/)

```python
# seabornを利用して描画する
# seabornの引数:hueは文字列である必要があるので
# 目的変数"Species"を文字列に変換
y2 = y
y2 = y2.replace({'Species': {0: "setosa"}})
y2 = y2.replace({'Species': {1: "virgiclor"}})
y2 = y2.replace({'Species': {2: "virginica"}})
# print(y2)

# データの結合
df2 = pd.concat([X, y2], axis=1)

# 描画
seaborn.pairplot(df2, hue="Species", palette="husl")

matplotlib.pyplot.show()
```

*４つの特徴量に関する相関係数行列を作る

```python
corr_mat = df2.corr(method='pearson')
```

```python
corr_mat
```

```python
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
plt.show()
```
