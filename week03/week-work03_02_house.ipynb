# 【問題4】欠損値への対応

* 各特徴量の欠損値の有無を確認する。

```python
plt.figure(figsize=(16,20)) #サイズ調整

sns.heatmap(df.isnull(), cbar=False); #ヒートマップ表示
```

* 欠損値が何パーセント含まれるかを確認する。

```python
#na_columns = df.isnull().sum()[df.isnull().sum()>0].sort_values()
print("missing_ration")
na_columns = (df.isnull().sum()[df.isnull().sum()>0].sort_values(ascending=False)) / 1460
na_columns
```

* 5つ以上欠損値がある特徴量（列）は削除する。

```python
# 削除対象の列

del_colum = df.isnull().sum()[df.isnull().sum()>4].sort_values()
print("---削除する項目---")
print(del_colum)
print("削除する項目数:",len(del_colum))
df_del = df.drop(del_colum.index, axis=1)

print(df.shape)
print(df_del.shape)
df_del
```

* 欠損値があるサンプル（行）は削除する。

```python
# 欠損値がある列を確認

na_colum2 = (df_del.isnull().sum()[df_del.isnull().sum()>0].sort_values())
na_colum2
```

```python
# 欠損値がある行を削除
df2 = df_del.dropna(subset=['Electrical'])

# 欠損値がないことを確認
df2.isnull().sum()[df2.isnull().sum()>0].sort_values()
```

# 【問題5】用語の調査

* 尖度（せんど)  
分布が正規分布からどれだけ尖っているかを表す統計量で、山の尖り度と裾の広がり度を示します。

* 歪度（わいど）
分布が正規分布からどれだけ歪んでいるかを表す統計量で、左右対称性を示す指標のことです。

# 【問題6】分布の確認

* 目的変数の分布を表示し、「尖度」と「歪度」も算出する

```python
# SalePrice - the property's sale price in dollars. 
# This is the target variable that you're trying to predict.

print("尖度", scipy.stats.kurtosis(df2.SalePrice))
print("歪度", scipy.stats.skew(df2.SalePrice))

sns.distplot(df2.SalePrice)
```

* 目的変数に対して対数変換を行う  
* 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
# 対数変換
np.log(df2.SalePrice)

print("尖度", scipy.stats.kurtosis(df2.SalePrice))
print("歪度", scipy.stats.skew(df2.SalePrice))

sns.distplot(df2.SalePrice)
```

* 対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。  
このように加工することで予測時の精度を上げることができます。

→対数変換前は正規分布になっていなかったが、対数変換後は正規分布に近づいた。

# 【問題7】相関係数の確認

* 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。  

→概要として家の品質、築年数、部屋数、バスルーム、ガレージなどの相関が高い。

```python
print(df2.shape)
print(df2.dtypes.value_counts())

corrmat = df2.corr()

# 注意:corr()メソッドではデータ型dtypeがobjectの列は除外される。
print(corrmat.shape)

#plt.subplots(figsize=(12,9))
#sns.heatmap(corrmat, vmax=0.9, square=True)

fig, ax = plt.subplots(figsize=(12, 9)) 
sns.heatmap(corrmat, square=True, vmax=1, vmin=-1, center=0)
```

* 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
high_corr_list = corrmat.sort_values("SalePrice", ascending=False)["SalePrice"][0:11]
print(high_corr_list)

# 相関係数が高いデータを抽出
df2_high_corr = df2[high_corr_list.index]

corrmat = df2_high_corr.corr()

fig, ax = plt.subplots(figsize=(12, 9)) 
#sns.heatmap(corrmat, square=True, fmt='.2f', vmax=1, vmin=-1, center=0)
sns.heatmap(corrmat, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10})
```

* 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

```python
high_corr_list.index
```

特徴量の説明

'OverallQual', Rates the overall material and finish of the house (素材/完成度のこと)  
'GrLivArea', Above grade (ground) living area square feet  
'GarageCars', Size of garage in car capacity  
'GarageArea', Size of garage in square feet  
'TotalBsmtSF', Total square feet of basement area  
'FullBath', full bathrooms above grade  
'1stFlrSF', First Floor square feet  
'YearBuilt', Original construction date  
'YearRemodAdd', Remodel date (same as construction date if no remodeling or additions)  
'TotRmsAbvGrd' Total rooms above grade (does not include bathrooms)  

* 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

```python
# 目的変数を抜く
high_corr_list2 = corrmat.sort_values("SalePrice", ascending=False)["SalePrice"][1:11]
print(high_corr_list2)

# 相関係数が高いデータを抽出
df2_high_corr2 = df2[high_corr_list2.index]

corrmat = df2_high_corr2.corr()

fig, ax = plt.subplots(figsize=(12, 9)) 
#sns.heatmap(corrmat, square=True, fmt='.2f', vmax=1, vmin=-1, center=0)
sns.heatmap(corrmat, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10})
```

```python
corrmat.where(corrmat > 0.7)
```

下記の相関係数が高い。  
(GarageArea, GarageCars)  
ガレージが広ければ広いほど車が置ける。

(1stFlrSF, TotalBsmtSF)
部屋の広さなので一致するため、相関がある。
TotalBsmSFが0の項目は、1stFlrSF, 2ndFlrSFの値を利用して補完する必要がある。

(TotRmsAbvGrd, GrLivArea)
Above Gradeにある部屋数と、Above Gradeの生活領域は比例する。


```python
plt.figure(figsize=(20, 15))

plt.subplot(2, 2, 1)
plt.scatter(df2["GarageArea"], df2["GarageCars"])
plt.xlabel("GarageArea(square feet)")
plt.ylabel("GarageCars(car capacity)")

plt.subplot(2, 2, 2)
plt.scatter(df2["1stFlrSF"], df2["TotalBsmtSF"])
plt.xlabel("1stFlrSF(square feet)")
plt.ylabel("TotalBsmtSF(square feet)")

plt.subplot(2, 2, 3)
plt.scatter(df2["TotRmsAbvGrd"], df2["GrLivArea"])
plt.xlabel("TotRmsAbvGrd")
plt.ylabel("GrLivArea(square feet)")

plt.show()
```
