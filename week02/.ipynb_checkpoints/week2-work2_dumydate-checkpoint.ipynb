# 【問題1】乱数の作成

* 平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$% <![CDATA[
\left[
\begin{array}{ccc}
  1.0 & 0.8\\
  0.8 & 1.0
\end{array}
\right] %]]>$

[numpy.random.multivariate_normal](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.multivariate_normal.html)

[共分散の説明](https://qiita.com/Seiji_Tanaka/items/5c8041dbd7da1510fbe9)

[あとで読む](https://qiita.com/skotaro/items/08dc0b8c5704c94eafb9)

```python
import numpy as np

# print文で省略しないように変更
np.set_printoptions(threshold=10000)
```

```python
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)

ndarray = np.random.multivariate_normal(mean, cov, 500)

#print(ndarray)

# 1列目を参照する場合
#print(ndarray[:, 0])
```

# 【問題2】散布図による可視化

[matplotlib](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
import matplotlib.pyplot as plt
```

```python
# データ準備
x = ndarray[:, 0]
y = ndarray[:, 1]

# グラフの装飾
plt.xlim(-6, 1) # (3)x軸の表示範囲
plt.ylim(-4, 4) # (4)y軸の表示範囲
plt.title("scatter", fontsize=20) # (5)タイトル
plt.xlabel("x", fontsize=20) # (6)x軸ラベル
plt.ylabel("y", fontsize=20) # (7)y軸ラベル
plt.grid(True) # (8)目盛線の表示
plt.tick_params(labelsize = 12) # (9)目盛線のラベルサイズ
 
# グラフの描画
plt.scatter(x, y, s=10, c="b",
                      marker="D", alpha=0.5) #(10)散布図の描画
plt.show()
```

# 【問題4】データの追加

新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)

ndarray2 = np.random.multivariate_normal(mean, cov, 500)

#print(ndarray)

# 1列目を参照する場合
#print(ndarray[:, 0])
```

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

[参考情報:レジェンド](https://qiita.com/matsui-k20xx/items/291400ed56a39ed63462)

```python
# データ準備
x = ndarray[:, 0]
y = ndarray[:, 1]

x2 = ndarray2[:, 0]
y2 = ndarray2[:, 1]

# グラフの装飾
plt.xlim(-6, 4) # (3)x軸の表示範囲
plt.ylim(-6, 4) # (4)y軸の表示範囲
plt.title("scatter", fontsize=20) # (5)タイトル
plt.xlabel("x", fontsize=20) # (6)x軸ラベル
plt.ylabel("y", fontsize=20) # (7)y軸ラベル
plt.grid(True) # (8)目盛線の表示
plt.tick_params(labelsize = 12) # (9)目盛線のラベルサイズ
 
# グラフの描画
plt.scatter(x, y, s=10, c="b", marker="D", label="0") #(10)散布図の描画
plt.scatter(x2, y2, s=10, c="r", marker="D", label="1") #(10)散布図の描画

plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=18)

plt.show()
```
